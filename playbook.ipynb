# Diffusion Load


```python

import torch
from diffusers import StableDiffusionPipeline

model_id = "CompVis/stable-diffusion-v1-4"
device = "cuda:0"

pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.bfloat16)
pipe = pipe.to(device)
```

```python
prompt = "space"
result = pipe(prompt)
result.images[0].show()
```
